```python
import os
```

```python
def check_file_for_characters(file_path, characters):
    try:
        with open(file_path, 'r') as file:
            file_content = file.read()
            for char in characters:
                if char in file_content:
                    return True
    except FileNotFoundError:
        print(f"文件 {file_path} 不存在")
    except PermissionError:
        print(f"无法访问文件 {file_path}")
    
    return False
```

```python
def find_tex_files(folder):
    tex_files = []
    
    # 遍历文件夹中的所有文件和子文件夹
    for root, dirs, files in os.walk(folder):
        for file in files:
            # 检查文件是否以 .tex 结尾
            if file.endswith('.tex'):
                # 构建文件的绝对路径
                file_path = os.path.join(root, file)
                # 判断是否有修改痕迹
                has_changes = check_file_for_characters(file_path, [r'\added{', r'\replaced{', r'\deleted{'])

                if has_changes:
                    # 将满足条件的文件添加到列表中
                    tex_files.append(file_path)

    return tex_files
```

```python
# 指定要搜索的文件夹路径
folder_path = '/Users/hwang/OneDrive/Documents/4_PhD/99_Thesis/data.code/DoctorThesis.tex/'

# 调用函数查找所有以 .tex 结尾的文件
result = find_tex_files(folder_path)

result
```

```python
for f in result:
    new_name = f[:]
    new_name = new_name[:-3] + "old.tex"
    print(new_name)
```

```python
import subprocess

def execute_cmd_command(command):
    try:
        # 执行 CMD 命令
        result = subprocess.run(command, shell=True, check=True, capture_output=True, text=True)
        # 输出命令执行结果
        print(f"命令输出：{result.stdout}")
        # 输出命令错误信息
        print(f"错误输出：{result.stderr}")
    except subprocess.CalledProcessError as e:
        print(f"命令执行出错：{e}")
```

```python
execute_cmd_command('ls .')
```
